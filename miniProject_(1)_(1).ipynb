```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
dataset = pd.read_csv('UpdatedResumeDataSet.csv')
ResumeDataset = dataset['Resume'].tolist()
```

```python
# User input - job description
job_description = """
We are looking for a software engineer with strong skills in Python and Java and Javascript And Jquery.
person must be sincere and hardworking.
"""
```

```python
# Function to clean and preprocess text
def preprocess_text(text):
    text = re.sub(r'\W', ' ', text.lower())
    return text

```

```python
# Function to calculate cosine similarity between job description and resumes
def calculate_similarity(job_description, resumes):
    processed_job_desc = preprocess_text(job_description)
    processed_resumes = [preprocess_text(resume) for resume in resumes]

    vectorizer = CountVectorizer().fit_transform([processed_job_desc] + processed_resumes)
    vectors = vectorizer.toarray()

    similarity_matrix = cosine_similarity(vectors)

    return similarity_matrix[0][1:]
```

```python
# Calculate similarity scores
similarity_scores = calculate_similarity(job_description, ResumeDataset)

# Recommend the best-matching resume
best_match_index = similarity_scores.argmax()
recommended_resume = ResumeDataset[best_match_index]
```

```python
# Display the result
print("Job Description:\n", job_description)
print("\nRecommended Resume:\n", recommended_resume)
print("\nSimilarity Score:", similarity_scores[best_match_index])
```
